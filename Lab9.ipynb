```r
before <- c(54, 25, 82, 76, 63, 82, 94, 72, 33, 90)
after <- c(68, 42, 80, 91, 70, 88, 90, 81, 38, 93)

# Sign test
# Null hypo: No change after the flexitime program i.e. P(+)=P(-)
# Alt hypo: There is an increase in attitude scores after the flexitime program i.e. P(+)>P(-)

sum(after-before>0)  # test statistic i.e. number of observations with '+'
1-pbinom(7, 10, 0.5)  # Under the null hypo, T~Bin(10, 0.5), p-value=P(T>=8)=1-P(T<=7)
# Conclusion: At 0.10 significance level, we reject the null hypo and conclude that there
# is an increase in attitude scores after the program.

# Wilcoxon Signed Rank test
# Null hypo: Distribution of scores before and after the program are identical i.e. P(+)=P(-)
# Alt hypo: Distribution of scores after > distribution of scores before the program i.e. P(+)>P(-)
wilcox.test(after, before, paired=T, alt='greater')
```

```r
# Friedman test
# Null hypo: Distribution of births across 4 seasons is identical
# Alt hypo: Distribution of births across 4 seasons is not all identical.

winter <- c(92, 9, 98, 19, 21, 58, 42)
spring <- c(112, 11, 109, 26, 22, 71, 49)
summer <- c(94, 10, 92, 19, 23, 51, 44)
fall <- c(77, 12, 81, 18, 24, 62, 41)
hos <- rep(c('A', 'B', 'C', 'D', 'E', 'F', 'G'), times=4)
my_data <- data.frame(hos, group=rep(c('Winter', 'Spring', 'Summer', 'Fall'), each=7), births=c(winter, spring, summer, fall))
my_data <- my_data[order(my_data$hos),]

rowrank <- NULL
# Compute row rank by subject
for (i in c('A', 'B', 'C', 'D', 'E', 'F', 'G')) {
    rowrank <- c(rowrank, rank(subset(my_data, my_data$hos==i)$births, ties.method='average'))
}

my_data <- data.frame(my_data, rowrank); my_data

ranksum <- aggregate(my_data$rowrank, by=list(my_data$group), FUN=sum)  # Compute rank sum by group
ranksum

# Calculate test statistic
n <- 7
k <- 4
A <- sum(my_data$rowrank^2)
B <- sum(ranksum[,2]^2)/n

t <- (k-1)*n*(B-n*k*(k+1)^2/4)/(A-n*k*(k+1)^2/4); t
p_value <- 1-pchisq(t, df=k-1); p_value

# Using friedman.test
friedman.test(my_data$births, my_data$group, my_data$hos)
```

```r
x <- c(59, 52, 42, 59, 24, 24, 40, 32, 63, 57, 36, 24)
y <- c(298, 303, 233, 287, 236, 245, 265, 233, 286, 290, 264, 239)
plot(x, y, xlab='Calcium (mg/100g)', ylab='Cholesterol (mg/100ml)')
# Check normality assumptions
shapiro.test(x)
shapiro.test(y)
# Conclusion: At 0.05 significance level, the data satisfies normality assumptions
var(x)
var(y)
cov(x,y)
r <- cov(x,y)/sqrt(var(x)*var(y))  # sample correlation coefficient(r)
r==cor(x,y) 
sqrt((1-r^2)/(length(x)-2))  # standard error of r

# Correlation test (t-test)
# Null hypo: correlation = 0
# Alt hypo: correlation != 0
cor.test(x,y)
# Conclusion: At 0.05 significance level, we reject the null hypo and conclude that
# correlation != 0

# Correlation test (F-test)
# Null hypo: correlation = 0
# Alt hypo: correlation > 0
f <- (1+abs(r))/(1-abs(r))
p_value <- 1-pf(f, 10, 10); p_value
# Conclusion: At 0.05 significance level, we reject the null hypo and conclude that
# correlation > 0

# Correlation test (Fisher's test)
# Null hypo: correlation = 0.5
# Alt hypo: correlation > 0.5
z <- (0.5*log((1+r)/(1-r))-0.5*log((1+0.5)/(1-0.5)))/(sqrt(1/(length(x)-3)))  # Z test statistic
p_value <- 1-pnorm(z); p_value
# Conclusion: At 0.05 significance level, we reject the null hypo and conclude that 
# correlation > 0.5
```

```r
# Confidence interval for correlation (Fisher Transform)
# 95% CI for Z=ln((1+rho)/(1-rho))
z_lower <- 0.5*log((1+r)/(1-r)) - qnorm(0.975)*sqrt(1/(length(x)-3))
z_upper <- 0.5*log((1+r)/(1-r)) + qnorm(0.975)*sqrt(1/(length(x)-3))
z_lower
z_upper

# 95% CI for rho using back-transform
rho_lower <- (exp(2*z_lower)-1)/(exp(2*z_lower)+1)
rho_upper <- (exp(2*z_upper)-1)/(exp(2*z_upper)+1)
rho_lower
rho_upper
```
