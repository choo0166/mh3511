```r
a <- c(102, 86, 98, 109, 92)
b <- c(81, 165, 97, 134, 92, 87, 114)
# Two independent samples t test with unequal variance
# Null hypo: mean of a = mean of b
# Alt hypo: mean of a < mean of b

abar <- mean(a)
bbar <- mean(b)
var_a <- var(a)
var_b <- var(b)
n_a <- length(a)
n_b <- length(b)

t_stat <- (abar-bbar)/sqrt(var_a/n_a + var_b/n_b); t_stat
v <- ((var_a/n_a) + (var_b/n_b))^2/((var_a/n_a)^2/(n_a-1)+(var_b/n_b)^2/(n_b-1))
v
p <- pt(t_stat, df=v); p

t.test(a, b, conf.level=0.9, var.equal=F, alt='less')
# Conclusion: At 0.1 significance level, there is insufficient evidence to conclude that mean of a < mean of b.
```

```r
fac <- read.table('faculty.txt', header=T, sep='')
str(fac)
head(fac)

# Using aggregate(data.frame=, by=list(x), FUN=) to summarise data by variable(s). 
# The by argument is a list of variables to group by. This must be a list even if there is only one variable.
# The FUN argument is the function which is applied to all columns (i.e., variables) in the grouped data.

aggregate(fac$male, by=list(fac$male), FUN=length)  # Aggregates counts by gender
boxplot(fac$salary~fac$male)
# Comment: Male professors have larger range and higher median salary than women

# Two independent samples t test
# Null hypo: Mean salaries of male professors = mean salaries of female professors
# Alt hypo: Mean salaries of male professors > mean salaries of female professors
var.test(fac$salary~fac$male)  # Check if variances are equal
# Comment: Variances are not equal
t.test(fac$salary~fac$male, var.equal=F, alt='greater')
# Conclusion: At 0.05 significance level, we conclude that the mean salaries of male professors > mean salaries of female professors.

fac.ap <- subset(fac, fac$rank=='Assistant')  # Subsets dataframe to APs
# Carry out two independent samples t test as above
var.test(fac.ap$salary~fac.ap$male)  # Comment: Variances are not equal
t.test(fac.ap$salary~fac.ap$male, var.equal=F, alt='greater')
# Comment: Same conclusion
```

```r
new <- c(2250, 2410, 2260, 2200, 2360, 2320, 2240, 2300, 2090)
old <- c(1920, 2020, 2060, 1960, 1960, 2140, 1980, 1940, 1790)
d <- new-old
# Paired samples t test
# Null hypo: Mean difference = 250
# Alt hypo: Mean difference > 250
t.test(d, mu=250, alt='greater')
# Conclusion: At 0.05 significance level, there is insufficient evidence to conclude that mean difference > 250.
```

```r
drugA <- c(4,5,4,3,2,4,3,4,4)
drugB <- c(6,8,4,5,4,6,5,8,6)
drugC <- c(6,7,6,6,7,5,6,5,5)
# Create a dataframe
my_data <- data.frame(group=rep(c('drugA', 'drugB', 'drugC'), each=9), pain=c(drugA, drugB, drugC))
my_data
aggregate(my_data$pain, by=list(my_data$group), FUN=summary)  # summary statistics of pain by group
aggregate(my_data$pain, by=list(my_data$group), FUN=var)  # sample variance

# One-Way ANOVA test for equality of means (assume population variances are equal)
# Null hypo: Mean pain levels are equal across all drug groups
# Alt hypo: Not all mean pain levels are equal
summary(aov(my_data$pain~my_data$group))
# Conclusion: At 0.05 significance level, we conclude that not all mean pain levels are equal

pairwise.t.test(my_data$pain, my_data$group, p.adjust.method='none')  # Perform pairwise comparisons
# Comment: From the table of p-values, mean pain level of drug A is different from drug B and C
# while drug B and C have the same mean pain levels.
```
