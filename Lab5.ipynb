```r
coffee <- c(1.95, 1.78, 2.10, 1.82, 1.73, 2.01, 1.83, 1.90, 2.05, 1.85, 1.96, 1.98, 1.79)
qqnorm(coffee)
qqline(coffee, col='blue')
shapiro.test(coffee)

n <- length(coffee) 
xbar <- mean(coffee) 
s <- sd(coffee) 

alpha <- 0.10 
z <- qnorm(1-alpha/2)  # Z critical value 
t <- qt(1-alpha/2, df=n-1)   # T critical value

zCI90p <- c(xbar-z*s/sqrt(n), xbar+z*s/sqrt(n)) 
print(paste("n=",n, "; xbar=", xbar, "; s=", s )) 
print(paste("alpha=",alpha, "; z=", z, "; t=", t )) 
print(paste("z90% CI= [", zCI90p[1], zCI90p[2], "]")) 

tCI90p <- c(xbar-t*s/sqrt(n), xbar+t*s/sqrt(n)) 
print(paste("t90% CI= [", tCI90p[1], tCI90p[2], "]"))

t.test(coffee, conf.level=0.9)

# Comment: Note that sigma is unknown, if we assume the data is normally distributed (or sample size is large)
# then we use the sample sd as sigma and the test statistic is ~N(0,1) distributed, otherwise it is t 
# distributed with n-1 dof. 

## 1-sided CI ##
zp <- qnorm(1-alpha)
tp <- qt(1-alpha, df=n-1)
zCI90p1 <- c(-Inf, xbar+zp*s/sqrt(n))
print(paste("z90% CI= [", zCI90p1[1], zCI90p1[2], "]")) 
tCI90p1 <- c(-Inf, xbar+tp*s/sqrt(n))
print(paste("t90% CI= [", tCI90p1[1], tCI90p1[2], "]"))

t.test(coffee, conf.level=0.9, alt='less')
```

```r
airquality <- read.csv('airquality.csv', header=TRUE, sep=",")
head(airquality)
wind89 <- subset(airquality$Wind, airquality$Month==8 | airquality$Month==9); wind89
qqnorm(wind89)
qqline(wind89, col='blue')

n <- length(wind89)
xbar <- mean(wind89)
s <- sd(wind89)

# Hypothesis test with null hypo: mu=10, alternative hypo: mu!=10
z <- (xbar-10)/(s/sqrt(n))
z_pvalue <- pnorm(z) + 1-pnorm(abs(z))  # P(abs(Z) >= abs(z)) = 2*(1-pnorm(abs(z)))
z_pvalue

t <- (xbar-10)/(s/sqrt(n))
t_pvalue <- pt(t, df=n-1) + 1-pt(abs(t), df=n-1)  # P(abs(T) >= abs(t))
t_pvalue

t.test(wind89, mu=10)
```

```r
n <- 25000
p0 <- 0.1
phat <- 2700/25000

# Hypothesis testing with null hypo: p0 = 0.1, alternative hypo: p0 > 0.1
z <- (phat-p0)/sqrt((0.1*0.9)/n)
z_pvalue <- 1-pnorm(z)  # P(Z >= z)
z_pvalue

prop.test(x=2700, n=25000, p=0.1, alt='greater')
```

```r
n <- c(10, 11, 20, 30)
# Hypothesis testing with null hypo: X~N(0,1), alternative hypo: X~N(mu1, 1), mu1 > 0
# Decision rule is reject null hypo when xbar >= 0.5, otherwise do not reject.
# Type 1 error(alpha) is probability of rejecting null hypo when it is true, i.e. P(xbar >= 0.5|H0 is true)
# Type 2 error(beta) is probability of not rejecting null hypo when it is false, i.e. P(xbar < 0.5|H1 is true)

alpha <- 1-pnorm(0.5/sqrt(1/n))  # xbar is ~N(0, 1/n) distributed
err1 <- data.frame(n, alpha)
colnames(err1) <- c('n', 'alpha'); err1

mu1 <- c(1, 1.1, 1.2, 1.3)
beta <- pnorm(0.5-mu1/sqrt(1/11))  # xbar is ~N(mu1, 1/n) distributed
err2 <- data.frame(mu1, 1-beta)  # Power is given by 1-beta
colnames(err2) <- c('mean', 'power'); err2
```
