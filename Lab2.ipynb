```r
func1 <- function(x){
    if (x < 100) {
        if (x < 10) {
            result <- 'single digit'
        }
        else {
            result <- 'double digits'
        }
    }
    else if (x <= 999) {
        result <- 'triple digits'
    }
    else {
        result <- 'more than 3 digits'
    }
    print(paste(x, 'is', result))
}

func1(5000)
```

```r
m2 <- function(x){
    y <- numeric(length(x))
    for (i in 1:length(x)){
        y[i] <- (x[i]-mean(x))^2
    }
    return(mean(y))
}
v <- seq(1, 10)
m2(v)

#install.packages('e1071')
library(e1071)
# Verify with moment function in e1071 package
moment(v, order=2, center=TRUE)
```
