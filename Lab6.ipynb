```r
set.seed(123)

# b: the number of samples that we want to code to run.
# n: the sample size of each sample.
# mu: the population mean, here we use an exponential distribution with mean=4, or rate=1/4.
# alpha: the value such that 100(1 âˆ’ alpha)% is the confidence level

CI_exp <- function(b, n, mu, alpha) {
    expsample <- matrix(rexp(n*b, 1/mu), ncol=n)
    xbar <- numeric(b)
    xsd <- numeric(b)
    mu_low <- numeric(b)
    mu_upper <- numeric(b)
    for (i in 1:b) {  # each row of expsample is a sample of size n from Exp(4) distribution
        xbar[i] <- mean(expsample[i,])
        xsd[i] <- sd(expsample[i,])
        # Construct 100(1 - alpha)% CI using normal distribution
        mu_low[i] <- xbar[i] - qnorm(1-alpha/2)*xsd[i]/sqrt(n)
        mu_upper[i] <- xbar[i] + qnorm(1-alpha/2)*xsd[i]/sqrt(n)
    }
    contains_mu <- ifelse(mu_low < 4 & mu_upper > 4, 1, 0)
    result <- rbind(mu_low, mu_upper, contains_mu)
    print(result)
    print(sum(contains_mu==1)/length(contains_mu))  # Prints proportion of CIs which contain mu
}

CI_exp(1000, 200, 4, 0.10)
```

```r
smoker <- read.table('smoker.txt', header=TRUE, sep=''); head(smoker)
summary(smoker)
smoketable <- table(smoker$Smoke, smoker$SES); smoketable
margin.table(smoketable)  # Returns total observations
margin.table(smoketable, 1)  # Returns row sums
margin.table(smoketable, 2)  # Returns col sums
```

```r
colsum <- matrix(colSums(smoketable), ncol=3)  # 1x3 matrix
rowsum <- matrix(rowSums(smoketable), ncol=1)  # 3x1 matrix
ex_smoketable <- rowsum %*% colsum / sum(colsum)  # Element-wise division
smoketable <- as.matrix(smoketable)
ex_smoketable  # table of expected values assuming row and column variables are independent
chisq_tstat <- sum((smoketable-ex_smoketable)^2/ex_smoketable); chisq_tstat
v <- (nrow(smoketable)-1)*(ncol(smoketable)-1); v
p_value <- 1-pchisq(chisq_tstat, df=v) ; p_value
summary(smoketable)
```

```r
lab <- matrix(c(16, 24, 654, 306), ncol=2, byrow=TRUE)
colnames(lab) <- c('Day', 'Evening')
rownames(lab) <- c('No', 'Yes')
lab
chisq.test(lab)  # At alpha level of 0.05, there is an association between conformance and shift
row_percent <- as.matrix(prop.table(lab, 1))*100; row_percent
col_percent <- as.matrix(prop.table(lab, 2))*100; col_percent

# Row percentage table answers the question 'Given the test is conforming/non-conforming, how likely is it to be performed in the day/evening?'
# So row percentage tables are appropriate if we believe the row variable to be explanatory and the column variable is the response.

# Column percentage table answers the question 'Given the test is performed in the day/evening, how likely is it to be conforming/non-conforming?'
# So column percentage tables are appropriate if we believe the column variable to be explanatory and the row variable is the response.

# In this case, the column percentage table seems more appropriate.
```
