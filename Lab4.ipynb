```r
### Sampling in R ###
U <- c(90, 95, 108, 103, 94, 109, 108, 93)  # A population of size 8, data assumed to be iid
mean(U)  # Population mean
ran_sample1 <- sample(U, size=4, replace=FALSE); ran_sample1  # Take a random sample of size 4 without replacement
allsamples <- combn(U, 4)  # Generates all possible samples of size 4 from U
head(allsamples)  # Each column represents a random sample
xbar <- colMeans(allsamples)  # A vector of sample means corresponding to each sample(column)
# Plot sampling distribution of xbar
hist(xbar, breaks=20)  # breaks argument controls amount of datapoints in each interval
```

```r
### Illustration of CI for unknown sigma^{2} and n<30 ###
expsizen <- function(n) {
    expsample <- matrix(rexp(n*1000, 1/3), ncol=n)  # Generates random values from Exp(3) distribution
    xbar <- numeric(1000)
    xsd <- numeric(1000)
    mu_low <- numeric(1000)
    mu_upp <- numeric(1000)
    for (i in 1:1000) {
        xbar[i] <- mean(expsample[i,])  # xbar stores sample mean of a sample of size n (corresponds to row i of matrix expsample)
        xsd[i] <- sd(expsample[i,])  # xsd stores sample sd 
        # Construct 90% CI using t-distribution
        mu_low[i] <- xbar[i] - qt(0.95, n-1)*xsd[i]/sqrt(n)
        mu_upp[i] <- xbar[i] + qt(0.95, n-1)*xsd[i]/sqrt(n)
        
    }
    contains_mu <- ifelse(mu_low < 3 & mu_upp > 3, 1, 0)  # Check if mu falls within the CI
    result <- rbind(mu_low, mu_upp, contains_mu)
    print(result)
    print(sum(contains_mu==1)/length(contains_mu))  # Proportion of CIs in which mu falls in 
}
expsizen(25)
```

```r
cfactor <- ifelse(roomwidth$unit=='metres', 3.28, 1)
# cfactor
roomwidth$convert <- roomwidth$width*cfactor
roomwidth
```

```r
boxplot(roomwidth$convert ~ roomwidth$unit)
mean(subset(roomwidth, roomwidth$unit=='metres')$convert)
sd(subset(roomwidth, roomwidth$unit=='metres')$convert)

mean(subset(roomwidth, roomwidth$unit=='feet')$convert)
sd(subset(roomwidth, roomwidth$unit=='feet')$convert)

# Comment: Since the true width is unknown, we cannot compare the 
# estimation based on mean. Hence, we chose the estimate with a smaller variance.
```

```r
t.test(subset(roomwidth, roomwidth$unit=='metres')$convert)
t.test(subset(roomwidth, roomwidth$unit=='feet')$convert)
```
