#### QQPlots (against Normal)

##### Quantiles
- With your data sorted in ascending order, quantiles are the various data points labelled as being the point below which a certain proportion of the data fall, e.g. the 0.5 quantile or $50^{th}$ percentile of the standard normal distribution is 0.
- A normal qqplot plots the $i^{th}$ order statistic of $X$ against the $\frac{i}{n+1}$ quantile of a normal distribution. (Detailed explanation at https://www.youtube.com/watch?v=X9_ISJ0YpGw)
- A straight line qqplot shows that the data is drawn from a normal distribution. Distributions with heavier tails (more area in the tails) will deviate more from the qqplot reference line at the tails of the distribution
![image](https://mgimond.github.io/ES218/Week05b_files/figure-html/unnamed-chunk-27-1.png)

```r
babies <- read.table('babiesl.data', header=TRUE, sep='');head(babies)
str(babies)
babies_0 <- subset(babies, (smoke==1 | smoke==0))
smoker <- subset(babies_0, smoke==1); dim(smoker)
non_smoker <- subset(babies_0, smoke==0); dim(non_smoker)
summary(smoker[,1])
summary(non_smoker[,1])
```

```r
par(mfrow=c(1,2))
hist(smoker[,1])
hist(non_smoker[,1])
```

```r
options(repr.plot.width=7, repr.plot.height=6)
par(mfrow=c(1,2))
boxplot(smoker[,1])
boxplot(non_smoker[,1])
# Comment: Boxplot gives a better comparison with shift in Q3 and Q1 clearly shown
```

```r
# Generating a normally distributed set of observations
# Let X follow a normal distribution, with mean=2 and var=9
q <- seq(0.02, 0.98, by=0.02)  # generate quantiles in increments of 0.02
x <- 2+3*qnorm(q)  # qnorm(q) is the z-score calculated from quantiles then transformed back to X
x
```

```r
options(repr.plot.width=7, repr.plot.height=4)
par(mfrow=c(1,2))
# qqnorm() generates QQplots
qqnorm(smoker[,1], main='Smoker')
qqline(smoker[,1], col='blue')  # adds reference line
qqnorm(non_smoker[,1], main='Non-smoker')
qqline(non_smoker[,1], col='red')
# Comment: Smoking group appears to be normally distributed, non-smoking group have heavier tails
# resulting in more deviation from the qqplot at the tails
```
